```python
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
for column in df.columns[:-1]:
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f'Boxplot of {column} by species')
    plt.show()
```

```python
for column in df.columns[:-1]:
    sns.violinplot(x='species', y=column, data=df)
    plt.title(f'Violin plot of {column} by species')
    plt.show()
```

```python
iris = sns.load_dataset('iris')
fig, axs = plt.subplots(2, 2)
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    sns.histplot(iris, x=feature, hue='species', multiple='stack', ax=axs[row, col])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
species_counts = df['species'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Iris Species Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df.drop(['species'], axis=1))
plt.title('Iris Dataset Line Graph')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Iris Dataset Scatter Plot')
plt.show()
```

```python
g = sns.catplot(x="species", y="sepal_length", hue="species", data=df, kind="violin")
plt.title('Iris Dataset Cat Plot')
plt.show()
```
